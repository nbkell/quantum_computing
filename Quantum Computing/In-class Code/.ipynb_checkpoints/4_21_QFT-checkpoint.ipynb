```python
import numpy as np 
from qiskit import *
from qiskit.quantum_info import Statevector
from qiskit.visualization import array_to_latex
from qiskit.visualization import plot_state_city, plot_bloch_multivector
```

```python
from numpy import pi 
def qft_circ(start):
    circ = QuantumCircuit(4)

    # level 0
    circ.h(3)
    circ.barrier()

    # level 1
    circ.cp(pi/2, 2, 3)
    circ.h(2)
    circ.barrier()

    # level 2 
    circ.cp(pi/4, 1, 3)
    circ.cp(pi/2, 1, 2) 
    circ.h(1)
    circ.barrier()
    
    # level 3 
    circ.cp(pi/8, 0, 3) 
    circ.cp(pi/4, 0, 2)
    circ.cp(pi/2, 0, 1)
    circ.h(0)
    circ.barrier()
    

    circ.swap(1,2)
    circ.swap(0,3)

    sv = Statevector(start) 
    sv = sv.evolve(circ)
    return sv, circ
```

```python
import numpy.linalg as linalg

start = np.random.rand(16)
start = start / linalg.norm(start)

answer1, circ = qft_circ(start)
answer2 = np.fft.fft(start, norm='ortho')
circ.draw('mpl')
```
